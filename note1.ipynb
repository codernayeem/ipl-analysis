```python
import pandas as pd, numpy as np, seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('ipl_matches.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
win_by_runs = df[ df['win_by_runs'] > 0 ]
```

```python
win_by_runs.head()
```

```python
plt.figure(figsize=(16, 6))
plt.plot('win_by_runs' , data=win_by_runs)
plt.title('Match Win by Runs')
plt.xlabel('Matches')
plt.ylabel('Runs')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Distribution Plot of Win by Runs')

sns.histplot(x='win_by_runs', bins=30, data=win_by_runs, ax=axes[0], color='g', edgecolor=None)
sns.kdeplot(x='win_by_runs', data=win_by_runs, ax=axes[1], color='g')

plt.show()
```

```python
toss_plus_win = df[ df['toss_winner'] == df['winner'] ]
```

```python
toss_plus_win.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Toss Winner wins by run and wicket')
axes[0].set_title('Toss Decision : Bat')
axes[1].set_title('Toss Decision : Field')

sns.kdeplot(x='win_by_runs', ax=axes[0], data=toss_plus_win[toss_plus_win['toss_decision'] == 'bat' ])
sns.kdeplot(x='win_by_wickets', ax=axes[1], data=toss_plus_win[toss_plus_win['toss_decision'] == 'field' ])

plt.show()
```

```python
rcb = df[ (df['team1'] == 'Royal Challengers Bangalore') | (df['team2'] == 'Royal Challengers Bangalore') ]
rcb.head()
```

```python
rcb.shape
```

```python
all_tosses = rcb['toss_winner'].value_counts()
rcb_toss_win_rate = all_tosses['Royal Challengers Bangalore'] / rcb.shape[0] * 100
rcb_toss_win_rate
```

```python
rcb_toss = rcb[rcb['toss_winner'] == 'Royal Challengers Bangalore']
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 7))
axes[0].set_title('RCB Toss Decision')
axes[1].set_title('RCB Win Rate')

y = rcb_toss['toss_decision'].value_counts()
axes[0].pie(np.flip(y.values), labels=np.flip(y.keys()), wedgeprops={'edgecolor': 'black'}, shadow=True, autopct='%1.1f%%')

toss_and_win_by_runs = rcb_toss['win_by_runs'].value_counts()[0]
toss_and_win_by_wicket = rcb_toss['win_by_wickets'].value_counts()[0]
axes[1].pie([toss_and_win_by_runs, toss_and_win_by_wicket], labels=['Toss : Bat', 'Toss : Field'], wedgeprops={'edgecolor': 'black'}, shadow=True, autopct='%1.1f%%')

plt.show()
```
